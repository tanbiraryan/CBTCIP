```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly .express as px
```

```python
data = pd.read_csv('Unemployment_Rate_upto_11_2020 - Unemployment_Rate_upto_11_2020.csv')
data
```

```python
# checking dataset information
data.info()
```

```python
# describing the dataset
data.describe()
```

```python
#check null/missing values
data.isnull().sum()
```

```python
# rename columns
data.columns = ['States','Date','Frequency','Estimated Unemployment Rate',
                'Estimated Employed','Estimated Labour Participation Rate',
                'Region','Longitude','Latitude']
data
```

```python
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Employed',hue='Region',data=data)
plt.show()
```

```python
# plotting histplot

plt.figure(figsize=(10,8))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate",hue='Region',data=data)
plt.show()
```

```python
# dashboard to analyze the unemployment rate of each indian state
# plotting sunburst

unemployment = data[['States','Region','Estimated Unemployment Rate']]
figure = px.sunburst(unemployment,path=['Region','States'],
                     values='Estimated Unemployment Rate',
                     width=700,height=600, color_continuous_scale='RdY1Gn',
                     title="Unemployment Rate in India")
figure.show()
```

```python
# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])
# Plotting time series data
plt.figure(figsize=(12, 6))
plt.title('Unemployment Rate Over Time')
sns.lineplot(x='Date', y='Estimated Unemployment Rate', hue='Region', data=data)
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate')
plt.show()
```
